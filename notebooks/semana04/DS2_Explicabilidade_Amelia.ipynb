```python
import json
import os

# [TAREFA] Mapeamento entre nomes t√©cnicos e nomes compreens√≠veis
mapeamento_amigavel = {
    "empresa_aerea": "Nome da Companhia A√©rea",
    "aerodromo_origem": "Aeroporto de Partida",
    "aerodromo_destino": "Aeroporto de Chegada",
    "hora_dia": "Hor√°rio do Voo (Hora)",
    "dia_semana": "Dia da Semana",
    "media_atraso_empresa": "Frequ√™ncia Hist√≥rica de Atrasos da Empresa",
    "media_atraso_origem": "√çndice de Atrasos no Aeroporto de Origem",
    "media_atraso_destino": "√çndice de Atrasos no Aeroporto de Destino",
    "codigo_tipo_linha": "Tipo de Opera√ß√£o",
    "janela_critica": "Per√≠odo de Pico no Tr√°fego"
}

os.makedirs('entregaveis_ds2', exist_ok=True)
with open('entregaveis_ds2/mapeamento_amigavel.json', 'w', encoding='utf-8') as f:
    json.dump(mapeamento_amigavel, f, ensure_ascii=False, indent=4)

print("‚úÖ Tarefa: Listagem e Mapeamento conclu√≠dos.")
```

```python
import pandas as pd
import numpy as np

def extrair_explicabilidade_global(pipeline, mapeamento):
    """
    [TAREFA] Identificar as vari√°veis mais importantes de forma global.
    """
    try:
        model = pipeline.named_steps['model']
        if hasattr(model, 'feature_importances_'):
            valores = model.feature_importances_
        else:
            valores = np.abs(model.coef_[0])

        df_global = pd.DataFrame({
            'fator_influencia': list(mapeamento.values())[:len(valores)],
            'importancia_relativa': valores
        }).sort_values(by='importancia_relativa', ascending=False)

        # [ENTREG√ÅVEL] Arquivo JSON com a import√¢ncia global
        df_global.to_json('entregaveis_ds2/importancia_global.json', orient='records', force_ascii=False, indent=4)
        return df_global
    except Exception as e:
        print(f"‚ö†Ô∏è Pipeline aguardando treinamento: {e}")

print("‚úÖ Tarefa: L√≥gica de Import√¢ncia Global pronta.")
```

## 2. Dicionário de Terminologia Técnica
Para garantir a compreensão de todos, usamos as seguintes definições:
- **EDA (Exploratory Data Analysis):** Análise Exploratória de Dados - processo de investigação inicial para descobrir padrões, anomalias e validar premissas através de estatísticas e visualizações.
- **Desbalanceamento de Classes:** Situação onde uma categoria da variável alvo é muito mais frequente que as outras, podendo causar viés no modelo. **[in [12]](https://github.com/LaboraDev/hackathon_one_flightOnTime/blob/main/notebooks/semana01/S01_Consolidado_ETL_EDA.ipynb)**
- **Cardinalidade:** Número de valores únicos em uma variável categórica (ex: 600+ aeroportos diferentes).
- **Sazonalidade:** Padrões que se repetem em períodos específicos (horas do dia, dias da semana, meses do ano).
- **Target Encoding:** Técnica de substituir categorias por estatísticas calculadas (como média de atraso por aeroporto).
- **Data Quality Report:** Relatório que documenta a completude, consistência e confiabilidade dos dados. **[in [31-39]](https://github.com/LaboraDev/hackathon_one_flightOnTime/blob/main/notebooks/semana01/S01_Consolidado_ETL_EDA.ipynb)**
- **Distribuição de Cauda Longa:** Fenômeno estatístico onde a maioria dos valores se concentra em uma extremidade e poucos valores formam uma "cauda" na outra extremidade.

## 5. Regras e decisões do projeto
Estabelecemos os seguintes pontos fundamentais durante a análise exploratória:

*   **Foco em Padrões Observáveis:** Como a coluna `codigojustificativa` é nula, concentramos a análise em padrões temporais e geográficos mensuráveis, abandonando a busca por causalidade explícita. **[in [27]](https://github.com/LaboraDev/hackathon_one_flightOnTime/blob/main/notebooks/semana01/S01_Consolidado_ETL_EDA.ipynb)**
*   **Identificação de Sazonalidade:** A análise revelou que horários de pico (tarde/noite) e períodos de férias concentram maior probabilidade de atrasos, direcionando a criação de features cíclicas na Semana 03. **[in [42-45]](https://github.com/LaboraDev/hackathon_one_flightOnTime/blob/main/notebooks/semana01/S01_Consolidado_ETL_EDA.ipynb)**
*   **Segmentação por Desempenho:** Companhias aéreas e aeroportos apresentam performance heterogênea, justificando o uso de *Target Encoding* para capturar o histórico de cada entidade. **[in [46-53]](https://github.com/LaboraDev/hackathon_one_flightOnTime/blob/main/notebooks/semana01/S01_Consolidado_ETL_EDA.ipynb)**
*   **Confirmação do Desbalanceamento:** A distribuição 84/16 (não-atrasado/atrasado) confirmou a necessidade de métricas como F1-Score e estratégias de balanceamento para a fase de modelagem.
*   **Validação da Limpeza:** A ausência de outliers temporais extremos após o tratamento da Semana 01 confirmou a eficácia dos filtros aplicados (remoção de diferenças >24h).

## 6. Respostas às Perguntas-Chave

### 6.1 Qual era o objetivo da EDA neste projeto?
O objetivo foi identificar padrões de comportamento nos dados (sazonalidade [In[42-45] e [56]](https://github.com/LaboraDev/hackathon_one_flightOnTime/blob/main/notebooks/semana01/S01_Consolidado_ETL_EDA.ipynb), concentração geográfica [In[51-53]](https://github.com/LaboraDev/hackathon_one_flightOnTime/blob/main/notebooks/semana01/S01_Consolidado_ETL_EDA.ipynb) e desempenho de companhias [In[46-49]](https://github.com/LaboraDev/hackathon_one_flightOnTime/blob/main/notebooks/semana01/S01_Consolidado_ETL_EDA.ipynb)) e validar se as premissas de tratamentos aplicadas na Semana 01 foram eficazes para remover ruídos que pudessem afetar o modelo de Machine Learning.

### 6.2 Como está distribuída a variável alvo? Existe desbalanceamento?
Sim, há um **desbalanceamento severo**. A classe majoritária (Antecipado/Pontual) representa cerca de 84% dos dados, enquanto os voos com atrasos superiores a 15 minutos representam apenas ~16%.  A classe de atrasos críticos (>240 min) é raríssima, representando 0,41% dos registros. [In[18]](https://github.com/LaboraDev/hackathon_one_flightOnTime/blob/main/notebooks/semana01/S01_Consolidado_ETL_EDA.ipynb)

### 6.3 Quais variáveis parecem mais relevantes?
As dimensões **Temporais** (`hora_dia` [In[44]](https://github.com/LaboraDev/hackathon_one_flightOnTime/blob/main/notebooks/semana01/S01_Consolidado_ETL_EDA.ipynb)
, `mes` [In[45]](https://github.com/LaboraDev/hackathon_one_flightOnTime/blob/main/notebooks/semana01/S01_Consolidado_ETL_EDA.ipynb)

) demonstraram maior variabilidade em relação ao atraso.  A localização do aeroporto e o horário de partida são os principais indicadores de problemas operacionais.
### 6.4 Como a EDA influenciou o Feature Engineering?
A exploração revelou que usar apenas a variável "data" não era suficiente para capturar os padrões de atrasos. Decidimos extrair informações adicionais como `hora`, `dia da semana` e criar métricas agregadas, como a "média histórica de atraso por aeroporto", para dar ao modelo um contexto que o dado isolado não forneceria.  [In[34]](https://github.com/LaboraDev/hackathon_one_flightOnTime/blob/main/notebooks/semana01/S01_Consolidado_ETL_EDA.ipynb)


### 6.5 Quais limitações foram identificadas na fase exploratria?
A principal limitação foi a falta de informações sobre causas. Como o campo `codigojustificativa` é nulo, a EDA confirmou que o modelo será "descritivo de padrões" e não "explicativo de causas" como clima ou manutenção. [In[22]](https://github.com/LaboraDev/hackathon_one_flightOnTime/blob/main/notebooks/semana01/S01_Consolidado_ETL_EDA.ipynb)

## 7. Análise de Riscos para Modelagem

| Risco | Severidade | Mitigation Strategy |
| :--- | :--- | :--- |
| **Viés da Classe Majoritária** | Alta | Aplicar métricas balanceadas (F1-Score) e `scale_pos_weight` no XGBoost. |
| **Alta Cardinalidade** | Média | Uso de *Target Encoding* para variáveis com 100+ categorias. |
| **Data Leakage** | Alta | Remoção rigorosa de features de "chegada" e "real" do conjunto X. |
| **Falta de Causalidade** | Média | Focar em padrões estatísticos e criar features de contexto histórico. |
